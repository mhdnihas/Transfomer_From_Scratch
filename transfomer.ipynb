```python
import torch
import torch.nn as nn
import math
```

```python
class MultiHeadAttention(nn.Module):
    def __init__(self,d_model,num_heads):
      super(nn.MultiheadAttention,self).__init__()
      self.d_model=d_model
      self.num_heads=num_heads
      self.d_k=d_model//num_heads

      self.W_q=nn.Linear(d_model,d_model)
      self.W_v=nn.Linear(d_model,d_model)
      self.W_k=nn.Linear(d_model,d_model)
      self.W_o=nn.Linear(d_model,d_model)


    def scaled_dot_product_attention(self,Q,K,V,mask=None):
        attention_score=torch.matmul(Q,K.traspose(-2,-1))//math.sqrt(self.d_k)

        att_prob=torch.softmax(attention_score,dim=-1)
        output=torch.matmul(att_prob,V)
        return output

    def split_head(self,x):
        batch_size,seq_length,d_model=x.shape()
        return x.view(batch_size,seq_length,self.num_heads,self.d_k).transpose(1,2)


    def compine_heads(self,x):
        batch_size,_,seq_length,d_k=x.size()
        return x.transpose(1,2).contgouse().view(batch_size,seq_length,self.num_heads)

    def forward(self,Q,K,V,mask=None):

        Q=self.split_head(self.W_q(Q))
        K=self.split_head(self.W_k(K))
        V=self.split_head(self.W_v(V))

        att_output=self.scaled_dot_product_attention(Q,K,V,mask)

        return self.compine_heads(att_output)


        
```
